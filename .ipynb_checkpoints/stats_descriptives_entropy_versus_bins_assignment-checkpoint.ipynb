```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## for random variables

# create Brownian noise
N = 1023
brownnoise = np.cumsum( np.sign(np.random.randn(N)) )

optimal_bin_width = 2 * stats.iqr(brownnoise) * (N)**(-1/3)
optimal_bins_count = int(np.ceil((max(brownnoise) - min(brownnoise)) / optimal_bin_width ))

nPerBin, bins = np.histogram(brownnoise, optimal_bins_count)
probs = nPerBin / sum(nPerBin)
# compute entropy
entro = -sum( probs * np.log2(probs + np.finfo(float).eps) )
print(f'Optimal bins count: {optimal_bins_count}')
print(f'Entropy: {entro}')

fig,ax = plt.subplots(2,1,figsize=(4,6))
ax[0].plot(brownnoise)
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Data value')
ax[0].set_title('Brownian noise')

ax[1].hist(brownnoise, optimal_bins_count)
ax[1].set_xlabel('Data value')
ax[1].set_ylabel('Counts')

plt.show()
```

```python
### now compute and plot entropy versus bins count

# number of bins
maxbins = 70

# initialize
entropy_versus_bins = np.zeros(maxbins)

for nbins in range(1, maxbins):
    # bin the data and convert to probability
    nPerBin, bins = np.histogram(brownnoise, nbins)
    probs = nPerBin / sum(nPerBin)
    # compute entropy
    entro = -sum( probs * np.log2(probs + np.finfo(float).eps) )
    entropy_versus_bins[nbins] = entro

# remove the 0th element as was unused
entropy_versus_bins = np.delete(entropy_versus_bins, 0)
print(f'Max entropy: {max(entropy_versus_bins)}')
bin_counts_with_max_entropy = np.argwhere(entropy_versus_bins == np.amax(entropy_versus_bins))
print(f'Bin counts having max entropy: {bin_counts_with_max_entropy.flatten().tolist()}')

plt.plot(entropy_versus_bins)
plt.show()
```
