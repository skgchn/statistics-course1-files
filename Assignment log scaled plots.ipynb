```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
  
# Creating vectors X and Y
x = np.linspace(0.001, 10, 100)
y1 = x
y2 = np.exp(x)
  
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='purple')

# Show the plot
plt.show()
plt.close()
```

```python
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='purple')

plt.yscale('log')

# Show the plot
plt.show()
plt.close()
```

```python
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(y1, x, color='blue')
plt.plot(y2, x, color='purple')

plt.xscale('log')

# Show the plot
plt.show()
plt.close()
```
