```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## create the data

n = 1000
threshold = 5 # threshold for cropping data

data = np.exp( np.random.randn(n) )
#print('values in close proximity or above of threshold before replacement: {}'.format(data[data>(threshold-0.5)]))
# ORIGINAL LINE:
data[data>threshold] = threshold + np.random.randn(sum(data>threshold))*.1

#In the above statement,
#  data>threshold returns a list of 1000 true and false. true elements are those where the value
#  was greater than threshold.
#  sum(data>threshold) is counting up the number of times the condition is true for the values in data.

### equivalent steps of the above original single line statement - start ###
#data_above_threshold = data[data>threshold]
#print('data above threshold: {}'.format(data_above_threshold))

#data_above_threshold_len = len(data_above_threshold);
#print('data_thresh_len: {}'.format(data_above_threshold_len))

#data_above_threshold_new_values = threshold + np.random.randn(data_above_threshold_len) * .1
#print('data_above_threshold_new_values: {}'.format(data_above_threshold_new_values))

#data[data>threshold] = data_above_threshold_new_values
#print('values in close proximity of threshold after replacement: {}'.format(data[data>(threshold-0.5)]))
### equivalent steps of the above original single line statement - end ###

# show histogram
plt.hist(data,30)
plt.title('Histogram')
plt.show()

# show violin plot
plt.violinplot(data)
plt.title('Violin')
plt.show()
```
