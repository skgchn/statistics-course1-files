```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## create the data

n = 1000
threshold1 = 5 # threshold for cropping data1
threshold2 = 10 # threshold for cropping data2
data1 = np.exp( np.random.randn(n) )
data2 = 5 + np.random.randn(n)

data1[data1>threshold1] = threshold1 + np.random.randn(sum(data1>threshold1))*.1
data1[data2>threshold2] = threshold2 + np.random.randn(sum(data2>threshold2))*.1


# show histogram
plt.hist(data1,30)
plt.title('Histogram data1')
plt.show()

# show histogram
plt.hist(data2,30)
plt.title('Histogram data2')
plt.show()
```

```python
from statsmodels.graphics.boxplots import violinplot
fig, ax = plt.subplots()
print(fig)
print(ax)
violinplot([data1], positions=[0], show_boxplot=False, side='right', ax=ax, plot_opts={'violin_fc':'C0'})
violinplot([data2], positions=[0], show_boxplot=False, side='left', ax=ax, plot_opts={'violin_fc':'C1'})
```

```python
import pandas as pd
import seaborn as sns
data = pd.DataFrame({'data1': data1, 'data2': data2})
data = data.melt(var_name = 'dataset', value_name = 'y')
data['x'] = 0
sns.violinplot(data = data, x = 'x', y = 'y', split = True, hue = 'dataset')
```
